```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#x_train = np.random.randint([1,1,1,1,1,1,1])
#y_train = np.random.randint([1,1,1,1,1,1,1])
x_train = np.array([5,5,5,4,7,2,3,4.5,4.2,3.5])
y_train = np.array([4,3,3,4,3,1,2,3.2,2.5,2.8])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(x_train.shape) # each blocks dimension
m = x_train.shape[0]
m
```

```python
plt.scatter(x_train, y_train, marker= 'x', c= 'r')
```

```python
w = 100 # setting w and b
b = 100 
```

```python
def predicts(x,w,b):
    f_array = np.zeros(m)
    for i in range(m):
        f = w * x[i] + b
        f_array[i] = f
    return f_array

predicted_outputs_ycap = predicts(x_train, w, b)
print(predicted_outputs_ycap)
```

```python
plt.plot(x_train, predicted_outputs_ycap, c='b') # plotting the prediction line
plt.scatter(x_train, y_train, marker='x', c='r')
```

```python
def computing_error(x,y,w,b):
    m = x_train.shape[0]
    error_sum = 0
    for i in range(m):
        f = w*x[i]+b
        error = f - y[i]
        error_sum += error**2
    return (1/(2*m))*error_sum

model_train = computing_error(x_train, y_train, w, b)
print(model_train)
```

```python
def compute_gradient(x,y,w,b):
    dj_dw = 0
    dj_db = 0
    m = x.shape[0]
    for i in range(m):
        dj_dw += (((w * x[i]) + b) - y[i]) * x[i]
        dj_db += ((w * x[i]) + b) - y[i]
    dj_dw = dj_dw/m
    dj_db = dj_db/m

    return dj_dw, dj_db

djdw,djdb = compute_gradient(x_train,y_train,w,b)
print(djdw,djdb)
```

```python
def gradient_descent(x,y,w,b,alpha): # gradient descent for w and b
    for i in range(1000):
        dj_dw, dj_db = compute_gradient(x,y,w,b)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db

    return w,b

w,b = gradient_descent(x_train,y_train,w,b,0.1)
print(w,b)
```
